# Week 7-4 select() 문

```python
# select() 문
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen('http://www.pythonscraping.com/pages/page3.html')
bs = BeautifulSoup(html, 'html.parser')

title=bs.find('h1')
print(title.get_text())

title1=bs.select_one('h1')
print(title1.get_text())

title2=bs.select('h1')
print(title2[0].get_text())
```

```python
table=bs.find('table').findAll('tr')
for tr in table:
    print("="*10)
    print(tr.get_text())
```

```python
# 띄어쓰기(직계자식 아니어도 됨)
table1 = bs.select('table tr')
for tr in table1:
    print("="*10)
    print(tr.get_text())
```

```python
table=bs.find('table').findAll('td')
for tr in table:
    print("="*10)
    print(tr.get_text())
```

```python
# 띄어쓰기(직계자식 아니어도 됨)
table1 = bs.select('table td')
for tr in table1:
    print("="*10)
    print(tr.get_text())
```

```python
# 직계 자식 태그 검색: > 사용
```

```python
table1 = bs.select('table > tr')
for tr in table1:
    print("="*10)
    print(tr.get_text())
```

```python
# 직계 자식이 아니므로 제대로 작동하지 않음
table2 = bs.select('table > td')
for tr in table2:
    print("="*10)
    print(tr.get_text())
```

```python
# 태그의 class 탐색 하기
```

```python
table=bs.find('table').findAll('tr',{'class':'gift'})
for tr in table:
    print("="*10)
    print(tr.get_text())
```

```python
# 태그의 class 탐색 하기, select() 사용 시 . 사용
table1=bs.select('table tr.gift')
for tr in table1:
    print("="*10)
    print(tr.get_text()) 
```

```python
# 태그의 id 탐색 시
table=bs.findAll('table',{'id':'giftList'})
for tr in table:
    print("="*10)
    print(tr.get_text())
```

```python
# 태그의 id 탐색 시, select() 사용 시 # 사용
table1=bs.select('table#giftList')
for tr in table1:
    print("="*10)
    print(tr.get_text()) 
```

```python
# 태그의 class와 id 동시 탐색 시
table=bs.find('table').findAll('tr',{'id':'gift2','class':'gift'})
for tr in table:
    print("="*10)
    print(tr.get_text())
```

```python
# 태그의 class와 id 동시 탐색 시, select 사용 시
table1=bs.select('table tr.gift#gift2')
for tr in table1:
    print("="*10)
    print(tr.get_text()) 
```
